##Оптимизатор Hyperopt

```python
#!pip install hyperopt
```

```python
#делаем импорт
import hyperopt
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

**Зададим пространство поиска гиперпараметров**

Я не знаю как реализовать несколько вариантов пространства, как это делали в RandomizedSearchCV.
Если реализовать список словарей и передать в Hyperopt, то ругается на дубликаты меток. Поэтому выберем наиболее лучшие из предыдущих оптимизаторов

```python
# зададим пространство поиска гиперпараметров
space={'penalty': hp.choice(label='penalty', options=['l1', 'l2']),
       'solver' : hp.choice(label='solver', options=['liblinear', 'saga']),
       'C': hp.choice(label='C', options=[0.01, 0.1, 0.3, 0.5, 0.7, 0.9, 1])
      }
```

```python
# зафиксируем random_state
random_state = 42
def hyperopt(space, cv=5, X=X_train, y=y_train, random_state=random_state):
    # функция получает комбинацию гиперпараметров в "params"
    space = {'penalty': space['penalty'], 
             'solver': space['solver'], 
             'C': space['C']
            }
  
    # задаём модели требуемые параметры  
    model=linear_model.LogisticRegression(**space, random_state=random_state, max_iter=50)


    # проводим кросс-валидацию  
    score = cross_val_score(estimator=model, X=X, y=y, 
                            scoring='f1', cv=cv, n_jobs=-1).mean()
    
    # метрику необходимо минимизировать, поэтому ставим знак минус
    return {'loss': -score.mean(), 'space': space, 'status': STATUS_OK}
```

```python
# начинаем подбор гиперпараметров
%%time

trials = Trials() # используется для логирования результатов

best=fmin(hyperopt, # наша функция 
          space=space, # пространство гиперпараметров
          algo=tpe.suggest, # алгоритм оптимизации, установлен по умолчанию, задавать необязательно
          max_evals=20, # максимальное количество итераций
          trials=trials, # логирование результатов
          rstate=np.random.RandomState(random_state)# фиксируем для повторяемости результата
         )
print("Наилучшие значения гиперпараметров {}".format(best))
```

```python
def df_results(hp_results):
    """
    Отображаем результаты hyperopt в формате DataFrame 

    :hp_results: результаты hyperop
    :return: pandas DataFrame
    """ 

    results = pd.DataFrame([{**x, **x['space']} for x in  hp_results])
    results.drop(labels=['space', 'status'], axis=1, inplace=True)
    results.sort_values(by=['loss'], ascending=True, inplace=True, ignore_index = True)
    return results
```

```python
results = df_results(trials.results)
print("Наилучшие значения гиперпараметров: ")
print("penalty: {}".format(results['penalty'][0]))
print("solver: {}".format(results['solver'][0]))
print("C: {}".format(results['C'][0]))
```

```python
# рассчитаем точность для тестовой выборки
model = linear_model.LogisticRegression(
    random_state=random_state, 
    penalty=results['penalty'][0],
    solver=results['solver'][0],
    C=results['C'][0],
    max_iter=50
)
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
print('f1_score на обучающем наборе: {:.2f}'.format(metrics.f1_score(y_train, y_train_pred)))
print("accuracy на тестовом наборе: {:.2f}".format(model.score(X_test, y_test)))
y_test_pred = model.predict(X_test)
print('f1_score на тестовом наборе: {:.2f}'.format(metrics.f1_score(y_test, y_test_pred)))
```

Видим, что метрика с лучшими значениями гиперпараметров, совпадает с метрикой, полученной при поиске гиперпараметров в оптимизаторе RandomizedSearchCV

##Оптимизатор Optuna

```python
# Устанавливаем библиотеку
#!pip install optuna
```

```python
import optuna
print("Версия Optuna: {}".format(optuna.__version__))
```

```python
# зафксируем random_state
random_state = 42
def optuna_lr(trial, cv=5, X=X_train, y=y_train, random_state=random_state):
  # задаем пространства поиска гиперпараметров
  penalty = trial.suggest_categorical("penalty", ['l1', 'l2'])
  solver = trial.suggest_categorical('solver', ['liblinear', 'saga'])
  C = trial.suggest_categorical('C', [0.01, 0.1, 0.3, 0.5, 0.7, 0.9, 1])
  
  # задаём модели требуемые параметры  
  model=linear_model.LogisticRegression(penalty = penalty,
                                        solver = solver,
                                        C = C,
                                        random_state=random_state,
                                        max_iter=50)

  #model.fit(X,y)
  #score = metrics.f1_score(y, model.predict(X))
  # проводим кросс-валидацию  
  score = cross_val_score(estimator=model, X=X, y=y, 
                          scoring='f1', cv=cv, n_jobs=-1).mean()

  return score
```

```python
%%time
# cоздаем объект исследования
# напрямую указываем, что нам необходимо максимизировать метрику direction="maximize"
study = optuna.create_study(direction="maximize")
# ищем лучшую комбинацию гиперпараметров n_trials раз
study.optimize(optuna_lr, n_trials=20)
```

```python
# выводим результаты на обучающей выборке
print("Наилучшие значения гиперпараметров {}".format(study.best_params))
print("f1_score на обучающем наборе: {:.2f}".format(study.best_value))
```

```python
# рассчитаем точность для тестовой выборки
model = linear_model.LogisticRegression(**study.best_params,
                                        random_state=random_state,
                                        max_iter=50)
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
print("accuracy на тестовом наборе: {:.2f}".format(model.score(X_test, y_test)))
y_test_pred = model.predict(X_test)
print('f1_score на тестовом наборе: {:.2f}'.format(metrics.f1_score(y_test, y_test_pred)))
```

##Оптимизатор Hyperopt

```python
#делаем импорт
import hyperopt
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
# зададим пространство поиска гиперпараметров
space={'n_estimators': hp.quniform('n_estimators', 80, 200, 30),
       'max_depth' : hp.quniform('max_depth', 20, 40, 2),
       'min_samples_leaf': hp.quniform('min_samples_leaf', 2, 10, 1)
      }
```

```python
# зафиксируем random_state
random_state = 42
def hyperopt(space, cv=5, X=X_train, y=y_train, random_state=random_state):
    # функция получает комбинацию гиперпараметров в "params"
    space = {'n_estimators': int(space['n_estimators']), 
             'max_depth': int(space['max_depth']), 
             'min_samples_leaf': int(space['min_samples_leaf'])
            }
  
    # задаём модели требуемые параметры  
    model=ensemble.RandomForestClassifier(**space, random_state=random_state)


    # проводим кросс-валидацию  
    score = cross_val_score(estimator=model, X=X, y=y, 
                            scoring='f1', cv=cv, n_jobs=-1).mean()
    
    # метрику необходимо минимизировать, поэтому ставим знак минус
    return {'loss': -score.mean(), 'space': space, 'status': STATUS_OK}
```

```python
# начинаем подбор гиперпараметров
%%time

trials = Trials() # используется для логирования результатов

best=fmin(hyperopt, # наша функция 
          space=space, # пространство гиперпараметров
          algo=tpe.suggest, # алгоритм оптимизации, установлен по умолчанию, задавать необязательно
          max_evals=20, # максимальное количество итераций
          trials=trials, # логирование результатов
          rstate=np.random.RandomState(random_state)# фиксируем для повторяемости результата
         )
print("Наилучшие значения гиперпараметров {}".format(best))
```

```python
# зафксируем random_state
random_state = 42
def optuna_rf(trial, cv=5, X=X_train, y=y_train, random_state=random_state):
  # задаем пространства поиска гиперпараметров
  n_estimators = trial.suggest_int('n_estimators', 80, 200, 30)
  max_depth = trial.suggest_int('max_depth', 20, 40, 2)
  min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 10, 1)
  
  # задаём модели требуемые параметры  
  model=ensemble.RandomForestClassifier(n_estimators=n_estimators,
                                        max_depth=max_depth,
                                        min_samples_leaf=min_samples_leaf,
                                        random_state=random_state)

  # проводим кросс-валидацию  
  score = cross_val_score(estimator=model, X=X, y=y, 
                          scoring='f1', cv=cv, n_jobs=-1).mean()

  return score
```

```python
%%time
# cоздаем объект исследования
# напрямую указываем, что нам необходимо максимизировать метрику direction="maximize"
study = optuna.create_study(direction="maximize")
# ищем лучшую комбинацию гиперпараметров n_trials раз
study.optimize(optuna_rf, n_trials=20)
```

```python
# выводим результаты на обучающей выборке
print("Наилучшие значения гиперпараметров {}".format(study.best_params))
print("f1_score на обучающем наборе: {:.2f}".format(study.best_value))
```
