# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


ВАРИАНТ 2

```python
answers['27'] = '27. Daniel Radcliffe, Rupert Grint'  # Ответ на Вопрос №27
```

```python
data27 = data[['cast']].copy()  # копируем data в data26
data27['cast'] = data27.cast.str.split('|')  # разбиваем строку по "|"
data27['cast'] = sorted(data27.cast)  # сортируем значения в алфавитном порядке в колонке cast построчно

# формируем новую колонку pairs с возможными парными комбинациями актеров с помощью itertools.combinations
data27['pairs'] = data27.cast.apply(lambda p: list(combinations(p, 2)))
pairs = Counter()  # создаем словарь типа Counter с данными о парах актеров
for s in data27.pairs:  # цикл для прокрутки строк data27.pairs
    pairs += Counter(s)  # заполняем данные контейнера используя Counter()
pairs.most_common(3)  # с помощью метода most_common выводим 3 наиболее часто встречающихся пары в порядке убывания
# из предложенных вариантов ответа выбираем пару актеров Daniel Radcliffe и Rupert Grint
```
